# Preprocessing

Supprimer les lignes identiques

```python
df = df[['Unité recommandée', 'Unité UCUM', "Libellés d\'édition (Reférentiel JDV ANS)"]].drop_duplicates()
```

Supprimer les lignes où les unités sont à NaN

```python
df = df[pd.notnull(df['Unité recommandée'])]
```

Renommer les colonnes

```python
df = df.rename(columns={'Libellés d\'édition (Reférentiel JDV ANS)' : 'Name'})
```

```python
df
```

# Concaténer les unités

```python
def concat_units(row):
    if row['Unité recommandée'] != row['Unité UCUM']:
        return row['Unité recommandée'] + "|" + row['Unité UCUM']
    else:
        return row['Unité recommandée']
    
df["Unit"] = df.apply(concat_units, axis=1)
df = df.drop(["Unité recommandée", 'Unité UCUM'], axis=1).reset_index()
```

```python
df = df.groupby('Name')['Unit'].agg(lambda x: '|'.join(x.unique())).reset_index()
```

```python
dict_unit = dict(zip(df['Name'], df['Unit'].str.split('|')))
```

```python
dict_unit
```

```python
nom_fichier = "bioloinc_units.json"

# Utilisez la méthode dump pour écrire le dictionnaire dans le fichier JSON
with open(nom_fichier, 'w') as fichier_json:
    json.dump(dict_unit, fichier_json)
```
