```python
def comparer_dictionnaires(ref, result):
    elements_non_communs = {}
    for cle, valeur in ref.items():
        if cle not in result or result[cle] != valeur:
            elements_non_communs[cle] = valeur
    return elements_non_communs


def tests_docs(ref_tests, results):
    '''
        ref_tests: list defined above
        results: list of dict of values found
    '''
    if len(ref_tests) != len(results):
        print('Not the same length')
        return False
    
    nb_equals = 0
    nb_total = 0
    for i in range(len(results)):
        nb_total += len(ref_tests[i])
        if results[i] == ref_tests[i]:
            nb_equals += len(results[i])
        else:
            print(f'######### Document of index {i} #########')
            elt_non_comm = comparer_dictionnaires(ref_tests[i], results[i])
            diff = len(ref_tests[i]) - len(elt_non_comm)
            nb_equals += diff
            print(f'Elements not found: {elt_non_comm}')
            
    percentage = (nb_equals / nb_total) * 100
    print(f'Percentage of success: {percentage}%')
    return True   
```
