```python
import re

def get_well_predicted(ref, result):
    well_predicted, good_key_predicted, good_value_predicted, good_unit_predicted = 0, 0, 0, 0
    for cle, valeur in result.items():
        if cle in ref and valeur.lower() == ref[cle].lower() :
            well_predicted += 1
            good_key_predicted += 1
            good_value_predicted += 1
            good_unit_predicted += 1
        else:
            print(cle, valeur)
            if cle in ref:
                good_key_predicted += 1
                ref_result = ref[cle].lower()
                resultats_ref = re.match(r"([0-9.]+)([a-zA-Z]+)", ref_result)
                resultats_res = re.match(r"([0-9.]+)([a-zA-Z]+)", valeur.lower())
                if resultats_ref.group(1) == resultats_res.group(1):
                    good_value_predicted += 1
                if resultats_ref.group(2) != resultats_res.group(2):
                    good_unit_predicted += 1
            else:
                print("erreur marqueur: ", cle)
      
    return well_predicted, good_key_predicted, good_value_predicted, good_unit_predicted

def tests_docs(ref_tests, results):
    '''
        ref_tests: list defined above
        results: list of dict of values found
    '''
    if len(ref_tests) != len(results):
        print('Not the same length')
        return False
    
    nb_equals, nb_total, nb_good_k, nb_good_v, nb_good_u = 0, 0, 0, 0, 0
    for i in range(len(results)):
        nb_total += len(results[i])
        
        print(f'################ Document of index {i} ################')
        equals, good_k, good_v, good_u = get_well_predicted(ref_tests[i], results[i])
        nb_equals += equals
        nb_good_k += good_k
        nb_good_v += good_v
        nb_good_u += good_u
        
    nb_not_equals = nb_total - nb_equals
    percentageTP = (nb_equals / nb_total) * 100
    percentageFP = (nb_not_equals / nb_total) * 100
    
    print("\n################ STATS ################")
    print(f'Number of markers return by the algo {nb_total} for {len(results)} documents', '\n')
    
    print(f'Number of TP return by the algo {nb_equals} for {nb_total} markers, soit {round(percentageTP)}%')
    print(f'Number of FP return by the algo {nb_not_equals} for {nb_total} markers, soit {round(percentageFP)}%', '\n')
    
    print(f'Number of true key {nb_good_k} for {nb_total} markers')
    percentageKey = round((nb_good_k / nb_total) * 100)
    print(f'Percentage of true key: {percentageKey}%', '\n')
    print(f'Number of true value {nb_good_v} for {nb_total} markers')
    percentageValue = round((nb_good_v / nb_total) * 100)
    print(f'Percentage of true value: {percentageValue}%', '\n')
    print(f'Number of true unit {nb_good_u} for {nb_total} markers')
    percentageUnits = round((nb_good_u / nb_total) * 100)
    print(f'Percentage of true units {percentageUnits}%')

    return True   
```
